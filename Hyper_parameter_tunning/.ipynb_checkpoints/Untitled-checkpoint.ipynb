```python
X_class = data_class.drop("class",axis = 1)
y_class = data_class.loc[:,'class']
X_train_class,X_test_class,y_train_class,y_test_class = train_test_split(X_class,y_class,test_size=0.2)
accuracy_initial_class = []
f1_initial_class = []
for i in range(1,10):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(X_train_class,y_train_class)
    y_pred_class = knn_model.predict(X_test_class)
    f1_initial_class.append(f1_score(y_test_class, y_pred_class,average='macro'))
    accuracy_initial_class.append(accuracy_score(y_test_class, y_pred_class))
    ConfusionMatrixDisplay.from_predictions(y_test_class,y_pred_class)
    print("\n "+"Number of neighbors = "+str(i)+"\n "+ classification_report(y_test_class, y_pred_class))
```

```python
crit_list = ['gini','entropy']
est_list = range(4,81,4)
accuracy_rf1 = []
accuracy_rf2 =[]
for crit in crit_list:
    for est in range(4,81,4):
        rf_class = RandomForestClassifier(n_estimators=est,criterion=crit)
        rf_class.fit(X_train_class,y_train_class)
        y_pred_rf = rf_class.predict(X_test_class)
        if crit == 'gini':
            accuracy_rf1.append(accuracy_score(y_test_class, y_pred_rf))
        else :
            accuracy_rf2.append(accuracy_score(y_test_class, y_pred_rf))
    if crit == 'gini':
        df_rf1 = pd.DataFrame({"nr_est":est_list,
                      "accuracy":accuracy_rf1})
    else:
        df_rf2 = pd.DataFrame({"nr_est":est_list,
                      "accuracy":accuracy_rf2})
        #table = pd.pivot_table(df_tree,values='accuracy',columns='max_features',index='depth',aggfunc=np.sum)
```

```python
plt.figure(figsize=(15,15))
plt.plot( 'nr_est', 'accuracy', data=df_rf1, label = 'gini')
plt.plot( 'nr_est', 'accuracy', data=df_rf2, label='entropy')
plt.xlabel("number of estimators")
plt.ylabel("accuracy")
plt.grid()
plt.legend()
plt.show()
```

```python
kern_list = ['poly', 'rbf', 'sigmoid']
c_list = range(1,11)
accuracy_svc1 = []
accuracy_svc2 =[]
accuracy_svc3 =[]
for kern in kern_list:
    for c in range(1,11):
        svc_class = SVC(C=c,kernel = kern)
        svc_class.fit(X_train_class,y_train_class)
        y_pred_svc = svc_class.predict(X_test_class)
        if kern == 'poly':
            accuracy_svc1.append(accuracy_score(y_test_class, y_pred_svc))
        elif kern == 'rbf' :
            accuracy_svc2.append(accuracy_score(y_test_class, y_pred_svc))
        elif kern == 'sigmoid':
            accuracy_svc3.append(accuracy_score(y_test_class, y_pred_svc))
    if kern == 'poly':
        df_svc1 = pd.DataFrame({"C":c_list,
                      "accuracy":accuracy_svc1})
    elif kern == 'rbf' :
        df_svc2 = pd.DataFrame({"C":c_list,
                      "accuracy":accuracy_svc2})
    elif kern == 'sigmoid':
        df_svc3 = pd.DataFrame({"C":c_list,
                      "accuracy":accuracy_svc3})
        
        #table = pd.pivot_table(df_tree,values='accuracy',columns='max_features',index='depth',aggfunc=np.sum)
```
